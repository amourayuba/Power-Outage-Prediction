```python
# import useful libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl 
mpl.rcParams["figure.dpi"] = 150
import seaborn as sns
import os

# enable copy on write (default in pandas 3.0)
pd.options.mode.copy_on_write = True
```

```python
def read_merged(weather, year):
    return pd.read_csv('../../merged/merged_{}_{}.csv.gz'.format(weather, year)).drop(columns=['Unnamed: 0'])
```

```python
lightning = read_merged('lightning',2020)
```

```python
lightning.head()
```

```python
outage = lightning[lightning['POWER_OUTAGE']]
no_outage = lightning[lightning['POWER_OUTAGE']==False]
```

```python
plt.hist(outage['TOTAL_COUNT'], density=True, log=True, bins=50, label='There is a power outage')
plt.hist(no_outage['TOTAL_COUNT'],  density=True, log=True, bins=50,label='No power outage', alpha=0.5)
#plt.hist(light[light['TOTAL_COUNT']<40]['TOTAL_COUNT'], density=True, bins=50, label='All', alpha=0.5)
#plt.xlim(0,200)
plt.xlabel('Number of lighting events')
plt.legend()
```

```python
plt.hist(np.log10(outage['TOTAL_COUNT']), density=True, bins=50, label='there is a power outage')
plt.hist(np.log10(no_outage['TOTAL_COUNT']),  density=True, bins=50,label='No power outage', alpha=0.5)
#plt.hist(light[light['TOTAL_COUNT']<40]['TOTAL_COUNT'], density=True, bins=50, label='All', alpha=0.5)
#plt.xlim(0,200)
plt.xlabel('Log number of lighting events')
plt.legend()
```
